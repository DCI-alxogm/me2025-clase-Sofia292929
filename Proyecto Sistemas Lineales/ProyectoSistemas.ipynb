```python
import pandas as pd
import matplotlib.pyplot as plt

# === 1. Cargar tu CSV de resultados ==========================
df = pd.read_csv("resultados_perfume_template.csv")

# Asegurar que inf/nan no rompan los gráficos
df = df.replace([float('inf'), float('-inf')], float('nan'))

print(df)

# === 2. Gráfica de soluciones por método ==========================
plt.figure(figsize=(10,6))
for xi in ['x1','x2','x3','x4']:
    plt.plot(df['metodo'], df[xi], marker='o', linewidth=2, label=xi)

plt.title("Soluciones del sistema por método")
plt.xlabel("Método")
plt.ylabel("Valor de x")
plt.legend()
plt.grid(True)
plt.show()


```
